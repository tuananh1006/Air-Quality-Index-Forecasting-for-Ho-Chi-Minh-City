# 2. Preprocess Column

```python
def preprocess_column_names(df):
    df.columns = df.columns.str.strip().str.lower().str.title().str.replace(' ', '_')
    return df

def preprocess_weather(df_weather):
    # Drop column không cần thiết
    df_weather = df_weather.drop(columns=[ 'App_Temp',"Revision_Version","Slp","Ts","Datetime",
                                          'Azimuth','Dewpt','Dhi','Dni','Elev_Angle','Ghi','H_Angle','Pod','Revision_Status','Snow','Solar_Rad','Vis','Wind_Dir','Wind_Gust_Spd','Weather','Start_Date','End_Date'])

    #Rename column
    df_weather = df_weather.rename(columns={
        'Timestamp_Local':'Local Time',
        'Clouds': 'Clouds',
        'Precip': 'Precipitation',
        'Pres': 'Pressure',
        'Rh': 'Relative Humidity',
        'Temp': 'Temperature',
        'Timestamp_Utc': 'UTC Time',
        'Uv': 'UV_Index',
        'Wind_Spd': 'Wind Speed'
    })
    df_weather = df_weather.sort_values(by='Local Time')
    df_weather['Local Time'] = pd.to_datetime(df_weather['Local Time'])
    df_weather.set_index('Local Time', inplace=True)
    return df_weather

def preprocess_air_quality(df_air_quality):
    df_air_quality=df_air_quality.drop(columns=[
        "Start_Date","End_Date","Ts","Datetime"
    ])
    df_air_quality=df_air_quality.rename(columns={
        'Timestamp_Local':'Local Time',
        'Timestamp_Utc':'UTC Time',
    })
    df_air_quality = df_air_quality.sort_values(by='Local Time')
    df_air_quality['Local Time'] = pd.to_datetime(df_air_quality['Local Time'])
    df_air_quality.set_index('Local Time', inplace=True)
    return df_air_quality

def preprocess(df,type):
  df = preprocess_column_names(df)
  if type == "weather":
    df = preprocess_weather(df)
  elif type == "air_quality":
    df = preprocess_air_quality(df)
  return df
air_quality_23_preprocess = preprocess(air_quality_23,type="air_quality")
weather_23_preprocess = preprocess(weather_23,type="weather")
air_quality_24_preprocess = preprocess(air_quality_24,type="air_quality")
weather_24_preprocess = preprocess(weather_24,type="weather")
air_quality_25_preprocess = preprocess(air_quality_25,type="air_quality")
weather_25_preprocess = preprocess(weather_25,type="weather")
display(air_quality_24_preprocess.head(5))
display(weather_24_preprocess.head(5))
```

# 4. Bo Sung Du Lieu

```python
# Function to group consecutive timestamps
def group_consecutive_timestamps_by_city(df_mismatched):
    missing_ranges_by_city = {}
    if not df_mismatched.empty:
        for city, group in df_mismatched.groupby('City_Name'):
            mismatched_times = group['Local Time'].sort_values()
            ranges = []
            if not mismatched_times.empty:
                current_start_time = mismatched_times.iloc[0]
                current_end_time = mismatched_times.iloc[0]

                for next_time in mismatched_times[1:]:
                    if (next_time == current_end_time + pd.Timedelta(hours=1)) and (next_time.date() == current_end_time.date()):
                        current_end_time = next_time
                    else:
                        ranges.append((current_start_time.strftime('%Y-%m-%d:%H'), current_end_time.strftime('%Y-%m-%d:%H')))
                        current_start_time = next_time
                        current_end_time = next_time

                ranges.append((current_start_time.strftime('%Y-%m-%d:%H'), current_end_time.strftime('%Y-%m-%d:%H')))
            missing_ranges_by_city[city] = ranges
    return missing_ranges_by_city
```

## 4.4 Fetch Data

### Function

```python
import logging
import requests
import time
from datetime import datetime, timedelta

SLEEP_TIME = 1 # Giảm thời gian chờ giữa các request vì số lượng request ít hơn
# --- Cấu hình Logging ---
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s [%(levelname)s] %(message)s",
    handlers=[logging.StreamHandler()]
)
```

```python
def fetch_data(base_url, api_key, city_name, start_date_str, end_date_str):
    """Hàm chung để gọi API và lấy dữ liệu."""
    params = {"city": city_name, "start_date": start_date_str, "end_date": end_date_str, "key": api_key}
    try:
        url = f"{base_url}?city={city_name}&country=VN&start_date={start_date_str}&end_date={end_date_str}&key={api_key}"
        print(url.split('key')[0])
        response = requests.get(url)
        response.raise_for_status()
        return response.json()
    except requests.exceptions.RequestException as e:
        logging.error(f"Lỗi API request cho {city_name} [{start_date_str} -> {end_date_str}]: {e}")
        return None
```

```python
def crawl_and_save_ranges(data_type, base_url, api_key, output_file, ranges_by_city):
    """
    Hàm crawl chính, phiên bản cuối cùng.
    - Bỏ qua các khoảng đã crawl.
    - Lọc dữ liệu theo đúng giờ.
    - Thêm log chi tiết sau mỗi lần lưu file thành công.
    """
    if not ranges_by_city:
        logging.info(f"✅ Không có dữ liệu để crawl cho {data_type}.")
        return

    # Đọc file CSV và tạo set các khoảng đã crawl
    done_pairs = set()
    if os.path.exists(output_file):
        try:
            done_data = pd.read_csv(output_file)
            if {"city_name", "start_date", "end_date"}.issubset(done_data.columns):
                done_pairs = set(zip(
                    done_data["city_name"].astype(str),
                    done_data["start_date"].astype(str),
                    done_data["end_date"].astype(str)
                ))
                logging.info(f"Đã tìm thấy {len(done_pairs)} khoảng đã được crawl trong file {output_file}.")
        except Exception as e:
            logging.warning(f"Không thể đọc file đã có: {e}. Sẽ tiến hành crawl mà không bỏ qua.")

    for city, ranges in ranges_by_city.items():
        logging.info(f"--- Bắt đầu xử lý {len(ranges)} khoảng {data_type} cho: {city} ---")

        for start_range_str, end_range_str in ranges:

            # Kiểm tra trước khi crawl
            key = (city, start_range_str, end_range_str)
            if key in done_pairs:
                logging.info(f"⏩ Bỏ qua {city} [{start_range_str} → {end_range_str}] (đã crawl trước đó)")
                continue

            try:
                start_dt = datetime.strptime(start_range_str, '%Y-%m-%d:%H')
                end_dt = datetime.strptime(end_range_str, '%Y-%m-%d:%H')

                if data_type == 'airquality':
                    api_start_date = start_dt.strftime('%Y-%m-%d')
                    api_end_date = (end_dt.date() + timedelta(days=1)).strftime('%Y-%m-%d')
                else: # 'weather'
                    api_start_date = (start_dt - timedelta(days=1)).strftime('%Y-%m-%d')
                    api_end_date = (end_dt + timedelta(days=1)).strftime('%Y-%m-%d')


            except ValueError as e:
                logging.error(f"Lỗi định dạng ngày giờ '{start_range_str}': {e}. Bỏ qua...")
                continue

            # Crawl và lọc dữ liệu
            json_data = fetch_data(base_url, api_key, city, api_start_date, api_end_date)

            if json_data and "data" in json_data and len(json_data["data"]) > 0:
                records_from_api = json_data["data"]
                filtered_records = []

                for r in records_from_api:
                    try:
                        record_dt = datetime.strptime(r['timestamp_local'], '%Y-%m-%dT%H:%M:%S')
                        if start_dt <= record_dt <= end_dt:
                            r["city_name"] = city
                            r["start_date"] = start_range_str
                            r["end_date"] = end_range_str
                            filtered_records.append(r)
                    except (ValueError, KeyError):
                        continue

                # Lưu dữ liệu đã lọc
                if filtered_records:
                    df_to_save = pd.DataFrame(filtered_records)
                    df_to_save.to_csv(
                        output_file,
                        mode="a",
                        header=not os.path.exists(output_file),
                        index=False
                    )
                    logging.info(f"  -> ✅ Đã lưu thành công {len(filtered_records)} bản ghi cho {city} [{start_range_str} -> {end_range_str}]")

                else:
                    logging.warning(f"  -> ⚠️ API trả về {len(records_from_api)} bản ghi, nhưng không có bản ghi nào phù hợp.")
            else:
                logging.warning(f"  -> ⚠️ Không tìm thấy dữ liệu.")

            time.sleep(SLEEP_TIME)
```

```python
# Load original and missing 2025 data
air_quality_25 = pd.read_csv(DATA_PATH + 'air_quality_city_2025.csv')
weather_25 = pd.read_csv(DATA_PATH + 'weather_city_2025.csv')
air_quality_missing_25 = pd.read_csv(DATA_PATH + 'air_quality_city_missing_2025.csv')
weather_missing_25 = pd.read_csv(DATA_PATH + 'weather_city_missing_2025.csv')

# Filter original air_quality_25 and weather_25 data to exclude October using 'timestamp_local'
air_quality_25_filtered_october = air_quality_25[pd.to_datetime(air_quality_25['timestamp_local']).dt.month != 10].copy()
weather_25_filtered_october = weather_25[pd.to_datetime(weather_25['timestamp_local']).dt.month != 10].copy()

# Merge filtered original data with missing data
air_quality_25_merged = pd.concat([air_quality_25_filtered_october, air_quality_missing_25], ignore_index=True)
weather_25_merged = pd.concat([weather_25_filtered_october, weather_missing_25], ignore_index=True)

# Re-apply preprocessing to the merged dataframes
air_quality_25_preprocess = preprocess(air_quality_25_merged, type="air_quality")
weather_25_preprocess = preprocess(weather_25_merged, type="weather")

# Print lengths of merged 2025 dataframes
print(len(air_quality_25_preprocess))
print(len(weather_25_preprocess))
```
