```python
import pandas as pd
import os

ROOT_PATH= "/content/drive/MyDrive/School"
DATA_PATH = os.path.join(ROOT_PATH,"AirQuality/preprocess_data/")
```

```python
air_quality_23 = pd.read_csv(DATA_PATH + 'air_quality_23_preprocess.csv')
air_quality_24 = pd.read_csv(DATA_PATH + 'air_quality_24_preprocess.csv')
air_quality_25 = pd.read_csv(DATA_PATH + 'air_quality_25_preprocess.csv')
weather_23 = pd.read_csv(DATA_PATH + 'weather_23_preprocess.csv')
weather_24 = pd.read_csv(DATA_PATH + 'weather_24_preprocess.csv')
weather_25 = pd.read_csv(DATA_PATH + 'weather_25_preprocess.csv')
xa_phuong_city_name_TPHCM=pd.read_csv(DATA_PATH + 'xa_phuong_city_name_TPHCM_preprocess.csv')
train_location = pd.read_csv(DATA_PATH + 'train_location.csv')
```

```python
df_23 = pd.merge(
    air_quality_23,
    weather_23,
    on=['Local Time', 'Train Name'],
    how='inner'
)
len(df_23)
```

```python
df_24=pd.merge(
    air_quality_24,
    weather_24,
    on=['Local Time', 'Train Name'],
    how='inner'
)
len(df_24)
```

```python
df_25=pd.merge(
    air_quality_25,
    weather_25,
    on=['Local Time', 'Train Name'],
    how='inner'
)
len(df_25)
```

```python
df=pd.concat([df_23,df_24,df_25])
len(df)
```

```python
print(len(air_quality_23))
print(len(air_quality_24))
print(len(air_quality_25))
sum = len(air_quality_23)+len(air_quality_24)+len(air_quality_25)
print(sum)
```

```python
df_merge=pd.merge(df,
                  train_location,
                  on=['Train Name'],
                  how='inner')
df_merge
```
